# 1. `What` is DeFi?

**Computer in the `financial` sky**

- **Blockchain** $\to$ the **computer in the sky**.
- **DeFi** $\to$ the set of **financial applications** running on the blockchain machine.

**Definition**

> Decentralized Finance (`DeFi`) is an **open** **digital** ecosystem where financial services are produced through **automated smart contracts** in order to eliminate **financial intermediation**.

`DeFi` inherits `Blockchain` properties:
- A set of **<mark>public</mark>, <mark>interoperable</mark> and <mark>autonomous protocols</mark>** which are **<mark>universally accessible</mark>**
- Developed, maintained and used by an **<mark>open pool of pseudonymous agents</mark>** rather than a set of unique legal entities.

# 4. The future of DeFi

The fundamental challenge

**Recall the binary information structure** 
- **Public (`on-chain`)** 
    - Verification at (almost) zero cost
    - Any activity on DeFi is public and can be contracted upon.
- **Private (`off-chain`)** 
    - Infinitely costly to verify
    - Pseudonymity: identity-related information cannot be contracted upon in DeFi

$\to$ $\exists~$ **<mark>limit</mark> on the <mark>contracting space</mark>** and the **scope of applications** for DeFi protocols.

**The smart contract challenge** 
(Gans, 2022)

## The new intermediaries

The solution to expanding contracting power is to introduce **new forms of intermediation**: **<mark>information bridges</mark>**

- Oracles
- Ramps
- Token issuers (linked to `off-chain`)

Trade-off: **verification vs contracting power**

### Oracles

**Definition**

Oracles provide the necessary **bridge to unify `off-chain` and `on-chain` worlds** 
- Service that **connects smart contracts** with **external data sources**. 

Oracles **expand the contracting space** of DeFi applications at the cost of **`on-chain` (public) verification**.

**Applications**

1. **Price Feeds**
   <br> Oracles fetch and deliver real-time price data for assets to DeFi platforms
   <br><u>ex:</u> *cryptocurrencies, commodities*
   <br><u>application:</u> *liquidation thresholds*
2. **External Event Verification**
   <br> Oracles confirm external events
   <br><u>ex:</u> *weather, politics, sports*
   <br><u>application:</u> *derivatives, prediction markets, insurance*
3. **Cross-Chain Communication**
    <br> Oracles facilitate interactions between different blockchains, making cross-chain DeFi products possible.

### Challenges

**The private and social costs of unverifiable information**

- **Centralization**
   <br> Economies of scale/scope 
   $\to$ <u>cf:</u> *Market power, Too-big-to-fail*
- **Inefficiency**
    <br> Oracles information rents
    $\to$ <u>cf:</u> *Credit rating agencies, market benchmarks*
- **Vulnerabilities**
    <br> Token redemption (stablecoin)
    $\to$ <u>cf:</u> *Runs on mutual funds and banks*
    <br> Oracle manipulation
    $\to$ <u>cf:</u> *Audit collusion, corruption*
- **Liabilities**
    <br> Legal challenges: who has what right?
    $\to$ <u>cf:</u> *Consumer protection, prospectus issuance*

**Opportunities for public interventions** $\to$ <u>cf:</u> traditional market failures in information markets
- Public provision
- Licensed actors
- Regulated oracle/token markets
    <br> $\to$ <u>cf:</u> MICA and GENIUS ACT

# Policy discussion

>Regulatory regimes built around intermediaries as regulated processors of transaction information may **fit poorly with a disintermediated market structure**.

WEF (2021)

## A new policy framework?

**Should DeFi receive a different policy treatment?**

↳ How do DeFi services differ in their treatment of **information frictions**?

**Claim:** Understanding the shift in information structure sheds light on 
- the scope of DeFi applications
- risks and inefficiencies (new and old)
- appropriate policy approaches (warranted and feasible actions)

**Elements of reflection**
- **Limits on policy enforcement power and information acquisition**
    - Computer in the sky: no shutting down, no liability, etc.
- **New targets**: validators, protocols, issuers and oracles.
- **Eligible proposals:** warranted & feasible 
    - **Warranted**
        <br> Several frictions can be best addressed by the private sector
        <br> Find cases with limits to the production of private solutions 
            <br> ↳ likely benefits from public support. 
    - **Feasible**
        <br> Subset of warranted actions satisfying the technological constraints of DeFi 

## Open policy issues

- How much **decentralisation** is good for the economy?
    <br> *Technology alone is no guarantee against centralisation of DeFi*
- How to enforce **liabilities** in a pseudonymous ecosystem? (<u>public-private issue</u>)
    <br> *Ensure AML-KYC standards*
- How to manage **unverifiable** information? (<u>public-private issue</u>)
    <br> *Embedding off chain information introduces a new information friction
    <br> Policing **oracles** and **ramps** $\to$ Recovering the value of traditional frameworks*
- How to think about **macro-prudential regimes** in absence of enforcement powers?
    <br> *Interconnected protocols - Misaligned incentives to audit - Contagion to the real economy*


# Conclusion

The main value proposition offered by DeFi is a **shift in the governance of information.** 
- Address social costs associated with financial intermediation
- Introduce its own set of social costs (new and old)

**For the crypto enthusiasts**

- Substituting automated protocols and pseudonymity for the confidentiality of financial intermediaries does not fully remove the fundamental information frictions at the heart of financial contracts

**For policymakers**
- Derive policy approaches that are both **warranted** and **feasible** under the technological constraints of DeFi 
- Investigate how the treatment of information differs under different DeFi protocols <br> $\to$ Design appropriate approaches for public support and intervention
